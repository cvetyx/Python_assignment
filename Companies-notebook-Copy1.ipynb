```python
import requests
import json

url = 'https://api.ycombinator.com/companies/export.json?callback=true'

resp = requests.get(url=url)
data = resp.text[15:-2]
y = json.loads(data)
for el in y:
    print(el['name'], el['batch'], el['description'], el['url'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.DataFrame(y)
print(df)
```
